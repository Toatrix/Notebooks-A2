Y nuestros costos a un segundo dataframe (desde excel)

```python
precios_df=pd.read_csv("Precios Restaurant.csv")
```

```python
precios_df.head(20)
```

## Operaciones en Pandas

Vamos creando una columan nueva de ganancias que será igual a precio menos costo de precios y costos

```python
precios_df["Precio"]=precios_df["Precio"].str.replace("$","").astype("float")
```

```python
precios_df.head(10)
```

```python
precios_df["Costo"]=precios_df["Costo"].str.replace("$","").astype("float")
```

```python
precios_df.head()
```

```python
precios_df["Ganancias"]=precios_df["Precio"]-precios_df["Costo"]#aqui se agrega la serie ganancia
```

```python
precios_df
```

Vamos tirando la columna de "hora" del restaruant

```python
rest_df=rest_df.drop("Hora de Cobro",axis=1)#metodo para eliminar fila o columna de un DF. Para q funcione necesita especifica si se borra una columna o una fila.
#por defecto está configurado para filas, el "1 " es para columna y el "0" para filas
```

```python
rest_df
```

Arma un nuevo dataframe con todos los platillos de más de 100 pesos de precio (usando filtros)

```python
precios100_df=precios_df[precios_df["Precio"]>100]
```

```python
precios100_df
```

Arma una nueva columna que clasifique platillos en alto/bajo margen, según si su ganancia es mayor a 100 pesos

```python
precios_df["Margen"]=np.where(precios_df["Ganancias"]>100,"Alto","Bajo")#np.where es una funcion de condicion np.where(condition, [T, F])
```

```python
precios_df
```

Cambiale el nombre a la columna de Atendió por Mesero en el df de Restaurant

```python
rest_df=rest_df.rename(columns={"Atendió":"Mesero"})#metodo 
```

```python
rest_df
```

## Merging o Vlookup

Vamos a juntar los 2 dataframes usando un left join

```python
restprecio_df=rest_df.merge(precios_df,on=["Producto"],how="left")
```

```python
restprecio_df.head(20)
```

Cuales son las diferencias entre merge de Python y Vlookup de Excel?

## Quitar duplicados

Crea un nueva Serie con la columna de "Mesero" - obten los valores únicos

```python
meseros=rest_df["Mesero"].drop_duplicates()#metodo
```

```python
meseros
```

## Tablas Dinámicas

Pandas también nos permite analizar datos mediante tablas dinámicas

Vamos creando una tabla dinámica que nos muestre las ventas por categoría

```python
ventaxcat=pd.pivot_table(restprecio_df,values="Precio",index="Categoria",aggfunc="sum")#index es como filas y Values pues valores en excel, precio se pone en valores como excel
```

```python
ventaxcat
```

Ahora una tabla dinámica que nos muestre las ventas por categoría y por tipo en la columna

```python
ventaxtipo=pd.pivot_table(restprecio_df,values="Precio",index="Categoria",columns="Tipo_x",aggfunc="sum")
```

```python
ventaxtipo
```

Crea una tabla dináica que muestre las ventas por tipo y categoría (filas), y por mesero (columnas)

```python
ventaxtipomesero_df=pd.pivot_table(restprecio_df,values="Precio",index=["Tipo_x","Categoria"],columns="Mesero",aggfunc="sum")
```

```python
ventaxtipomesero_df
```

Crea una tabla dinámica que muestre las ventas promedio por mesero

```python
meseromean_df=pd.pivot_table(restprecio_df,"Precio",index=["Mesero"])#por default trae mesero 
```

```python
meseromean_df
```
